```python
from pliar.main import main
from pathlib import Path

model_dir = Path("..") / "data" / "paper_models"
if (
    not model_dir.exists()
    or sum(1 if p.is_file() else 0 for p in model_dir.iterdir()) == 0
):
    raise ValueError("No models to evaluate")
outdir = Path("..") / "data" / "processed" / "alignment"


commnands = []
for model in ("cgnn", "cgnn3d", "dimenet"):
    for split in ["train", "test"]:
        for split_type in ["scaffold", "pocket"]:
            main(
                model_dir / f"{model}_{split_type}_{split}_clean.csv",
                model_dir / f"{model}_{split_type}_{split}_masked.csv",
                outdir=outdir,
                prefix=f"{model}_{split_type}_{split}_",
            )
```
